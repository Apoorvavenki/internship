```python
import pandas as pd
path="/Users/apoorva/Desktop/untitled folder 4/train_dataset.csv"
data=pd.read_csv(path)
print(data)
print(data.info())
```

```python
inputs=data.drop(['Serial No','microorganisms'],axis=1)
print("inputs",inputs)
outputs=data['microorganisms']
print(outputs)
```

```python
import numpy as np
import sklearn 
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(inputs,outputs,train_size=0.8)
print(x_train)
print(x_test)
print(y_train)
print(y_test)
import sklearn 
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier(n_neighbors=13)
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
print(y_pred)
print(y_test)
```

```python
import sklearn 
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
print(cm)
```
